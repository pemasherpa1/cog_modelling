```python

%cd C:\Users\pemas\OneDrive\Desktop\baseline model\experiment_code_held2022-main\analysis\ana06

import pickle
import matplotlib.pyplot as plt

# Import saved variables for baseline data
with open('mismatch_data_baseline.p', 'rb') as f:
    baseline_data = pickle.load(f)
    
# Import saved variables for support system data
with open('mismatch_data.p', 'rb') as f:
    support_system_data = pickle.load(f)

# Plot both curves together
plt.plot(baseline_data['sorted_timestamps'], baseline_data['sorted_counts'], label='Baseline Data')
plt.plot(support_system_data['sorted_timestamps'], support_system_data['sorted_counts'], label='Support System Data')
plt.xlabel('Timestamp')
plt.ylabel('Mismatch')
plt.title('Mismatch Pattern Comparison')
plt.legend()
plt.show()
```
