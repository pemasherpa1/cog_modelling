```python

""""""

import pandas as pd
import pickle
import glob
import numpy as np
import os



# this line will ensure were on the same directory as the python source files
%cd C:\Users\pemas\OneDrive\Desktop\University\Master UOL\Research Project\Code Versions\with support mechanism nr1\Experiment Code\PS_experiment code\experiment_code_held2022-main


PATH_IN = 'analysis/ana01/'
PATH_OUT = 'analysis/ana02/'
CONDITIONS = ['2backPS']
SRATE = 0.05

print(os.getcwd())

#-------------------------------------------------------------------------------------

def calc_deviation(data: str, threshold: float):
    """Calculate average lane deviation."""
    lanepos = np.array(data['lanepos']).astype(float)

    # Filter out outliers
    filtered_lanepos = lanepos[np.abs(lanepos - np.mean(lanepos)) < threshold]

    deviation = np.abs(filtered_lanepos - 2.5) * 2.5
    deviation = np.mean(deviation)

    return deviation


#-------------------------------------------------------------------------------------

def calc_sdlp(data: str, threshold: float):
    """Calculate standard deviation of the lateral position."""
    lanepos = np.array(data['lanepos']).astype(float)

    # Filter out outliers
    filtered_lanepos = lanepos[np.abs(lanepos - np.mean(lanepos)) < threshold]

    sdlp = np.std(filtered_lanepos)

    return sdlp

if __name__ == '__main__':
    # Define outlier threshold
    outlier_threshold = 1                                     # Adjust here 

    # Load and process data
    files_condition = pickle.load(open(PATH_IN + 'files_sorted_behavior.p', 'rb'))

    # Initialize dictionaries
    deviation_all = dict.fromkeys(CONDITIONS)
    deviation_avg = dict.fromkeys(CONDITIONS)
    sdlp_all = dict.fromkeys(CONDITIONS)
    sdlp_avg = dict.fromkeys(CONDITIONS)

    for condition in CONDITIONS:
        deviation_condition = [] 
        sdlp_condition = []

        for filename in files_condition[condition]:
            data = pickle.load(open(filename, 'rb'))

            # Calculate average lane deviation per run with outlier filtering
            deviation = calc_deviation(data, outlier_threshold)
            deviation_condition.append(deviation)

            # Calculate standard deviation of lateral position with outlier filtering
            sdlp = calc_sdlp(data, outlier_threshold)
            sdlp_condition.append(sdlp)

        deviation_all[condition] = deviation_condition
        deviation_avg[condition] = np.mean(deviation_condition)
        sdlp_all[condition] = sdlp_condition
        sdlp_avg[condition] = np.mean(sdlp_condition)

        print(f"\n{condition} Model - Calculation of lane deviation and STD of lateral position done!\n")
        print(f" Thresholded at {outlier_threshold}")





    # Dump all lane deviation and SDLP into one file
    pickle.dump(deviation_all, open(PATH_OUT + 'lanedev_all_thresholded.p', 'wb'))
    pickle.dump(sdlp_all, open(PATH_OUT + 'sdlp_all_thresholded.p', 'wb'))

    # Save to CSV
    pd.DataFrame.from_dict(data=deviation_avg, orient='index').to_csv(PATH_OUT + 'lanedev_thresholded.csv', header=False)
    pd.DataFrame.from_dict(data=sdlp_avg, orient='index').to_csv(PATH_OUT + 'sdlp_thresholded.csv', header=False)
```
